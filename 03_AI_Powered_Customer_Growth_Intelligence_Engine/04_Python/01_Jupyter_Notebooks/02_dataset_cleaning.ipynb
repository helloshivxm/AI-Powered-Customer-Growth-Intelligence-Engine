```python
import pandas as pd
import numpy as np

sheet_1 = pd.read_excel("../../01_Raw_Data/online_retail_II.xlsx",
              sheet_name= "Year 2009-2010")

sheet_2 = pd.read_excel("../../01_Raw_Data/online_retail_II.xlsx",
              sheet_name= "Year 2010-2011")

retail_df = pd.concat(
    [sheet_1, sheet_2],
    ignore_index= True
)

print(retail_df.shape)
```

```python
retail_df = retail_df[ 
    ~retail_df["Invoice"].astype(str).str.startswith("C", na=False)
]
retail_df.shape
```
